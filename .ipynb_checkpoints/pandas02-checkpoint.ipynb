```python
import pandas as pd
data = {'state': ['Ohio', 'Ohio', 'Ohio', 'Nevada', 'Nevada', 'Nevada'],
        'year': [2000, 2001, 2002, 2001, 2002, 2003],
        'pop': [1.5, 1.7, 3.6, 2.4, 2.9, 3.2]}
frame = pd.DataFrame(data)

frame
# display the top five rows
frame.head()
```

```python
# customize the col name
pd.DataFrame(data, columns=['year','state','pop'])
```

```python
frame2 = pd.DataFrame(data, columns=['year','state','pop','tmp'],index=['one', 'two', 'three', 'four',
   ....:                              'five', 'six'])
frame2
```

```python
# get a specific cols
frame2['state']
```

```python
frame2.year
```

```python
# locate on one specific row
frame2.loc['three']
```

```python
# fill the col
frame2['tmp'] = 19
frame2
```

```python
# if new cols, create
frame2['newCol'] = frame2.state == 'Ohio'
frame2
```

```python
# delete specific cols
del frame2['newCol']
frame2
```

```python
# 对于两层嵌套，最外层是列名，里面的是行名
pop = {'Nevada': {2001: 2.4, 2002: 2.9},
....:        'Ohio': {2000: 1.5, 2001: 1.7, 2002: 3.6}}
frame3 = pd.DataFrame(pop)
frame3
```

```python
# DataFrame 的转置
frame3.T
```

```python
# specify the rows
pd.DataFrame(pop, index=[2001, 2002, 2003])
```

```python
# setting the col name & row name
frame3.index.name = 'year'; frame3.columns.name = 'state'
frame3
```

```python
# the values of frame will return into the array
frame3.values
```
